## 4. Univariate Analysis

Business framing:  

Hotels care about trends like average stay length, customer mix, and pricing. A good EDA starts with understanding single variables and their distribution.

- Select at least 3 individual variables to explore
- Use plots and summary methods (e.g. info(), describe(), etc)  to describe the distribution (hint: we are only looking at the values of one variable, so think of plots you've used in the past that DON'T compare 2 variables.)
- Focus on what matters from a business standpoint (e.g., pricing, cancellations, guest types)

```python
# Your code for univariate analysis (e.g., plots, value counts) üîß
# Variable 1
print(df['adr'].describe())

plt.hist(df['adr'], bins=50, color='green', edgecolor='black')
plt.title('Distribution of Average Daily Rate (ADR)')
plt.xlabel('ADR')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
#Variable 2
# Summary counts
print(df['is_canceled'].value_counts())
print(df['is_canceled'].value_counts(normalize=True))  # % of cancellations

# Bar plot
df['is_canceled'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Booking Cancellations')
plt.xlabel('Canceled (1) vs Not Canceled (0)')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# Variable 3
# Create total stay variable
df['total_stay'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']

# Summary
print(df['total_stay'].describe())

# Histogram
plt.hist(df['total_stay'], bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Total Length of Stay')
plt.xlabel('Total Nights Stayed')
plt.ylabel('Number of Bookings')
plt.show()
```

### In your markdown:
1. Variable 1 ‚Äì I explored the number of booking v.s. average daily rate, or amount charged per day. I found that when the ADR is higher, there are fewer bookings, the drop off is quite drastic at high levels.
2. Variable 2 ‚Äì I explored the count of bookings and whether customers cancelled. I found that a significant portion of bookings are cancelled, although more are not cancelled than cancelled.
3. Variable 3 ‚Äì I compared number of bookings to toal nights stayed. I found that the majority of bookings are short term and that there is a trend where total nights stayed increases when number of bookings decreases.

## 5. Bivariate Analysis

Business framing:  

Stakeholders often ask: ‚ÄúWhat drives cancellations?‚Äù or ‚ÄúDo longer stays mean higher revenue?‚Äù Bivariate analysis helps you uncover those kinds of relationships.

- Choose 2 relevant variable pairs (e.g., `lead_time` vs. `is_canceled`, or `adr` vs. `customer_type`)
- Use scatterplots, grouped bar plots, or boxplots to explore the relationships
- Interpret what these relationships could mean for the hotel business

```python
# Your code to analyze variable relationships (e.g., scatterplots, grouped bars) üîß
# Scatterplot: lead_time vs is_cancelled
sns.barplot(x='is_canceled', y='lead_time', data=df)
plt.title("Lead Time vs Cancelation Rates")
plt.xlabel("Canceled")
plt.ylabel("Lead Time")
plt.show()
```

```python
# ADR vs Customer Type
sns.barplot(data=df, x='adr', y='customer_type')
plt.title("Average Room Rate Based On Customer Type")
plt.xlabel("adr")
plt.ylabel("customer type")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
